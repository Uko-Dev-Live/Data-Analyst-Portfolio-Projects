## Handling Missing Data

```python
# Finding Missing data
for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, percent_missing))
```

### Normalizing the Date 

```python
# Cleaning and Filling the missing data in released column with data from year column from same row 

yearList = []
for char, char1, char2 in zip(df['released'].isna(), df['year'].astype(str), df['released'].astype(str)):
    if char == True:
        #print(char1)
        yearList.append(char1)
        #print(char2)
    for elem in char2:
        if elem == '(':
            yearList.append(char2[char2.index(elem)-5:char2.index(elem)])
    

df['yearcorrected'] = yearList

len(yearList)
df['yearcorrected'].str.strip().head()
```

```python
# Some values in the has spaces before them which can be removed using .str.strip() method 
df['yearcorrected'] = df['yearcorrected'].astype(str).str.strip()
df['yearcorrected'].head()
```

#### correlation

```python
# scattered plot budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.ylabel('Budget for Company')
plt.xlabel('Gross Earnings')
plt.show
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color': 'blue'} , line_kws={'color':'black'})
```

```python
# How  much is the correlation

df.corr(method='pearson') # Pearson, Kendall, Spearman
```

```python
correlation_matrix = df.corr(method='pearson') # Pearson
sns.heatmap(correlation_matrix, annot=True)


plt.title('Correlation Metric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show
```

```python
# convert object to numeric values  
df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()        
```

```python
correlation_matrix = df_numerized.corr(method='pearson') # Pearson
sns.heatmap(correlation_matrix, annot=True)


plt.title('Correlation Metric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show
```
